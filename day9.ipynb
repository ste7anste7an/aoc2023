```python
r="""0 3 6 9 12 15
1 3 6 10 15 21
10 13 16 21 30 45"""

r=open('day9_inp.txt','r').read()
r=r.split('\n')

import numpy as np
tot=0
for s in r:
    sum=0
    seq=np.array([int(i) for i in s.split()])
    while seq.any():
        sum+=seq[-1]
        seq=np.diff(seq)
        #print(seq)
    tot+=sum   
        
print(tot)
```

```python
r="""0 3 6 9 12 15
1 3 6 10 15 21
10 13 16 21 30 45"""

r=open('day9_inp.txt','r').read()
r=r.split('\n')

import numpy as np
tot=0
for s in r:
    sums=[]
    seq=np.array([int(i) for i in s.split()])
    while seq.any():
        sums.append(seq[0])
        seq=np.diff(seq)
    
    print(sums)   
    sum=0
    for ss in sums[::-1]:
        sum=ss-sum
    print(sum)
    tot+=sum
print(tot)
```
