```python
from aocd import get_data
r="""R 6 (#70c710)
D 5 (#0dc571)
L 2 (#5713f0)
D 2 (#d2c081)
R 2 (#59c680)
D 2 (#411b91)
L 5 (#8ceee2)
U 2 (#caa173)
L 1 (#1b58a2)
U 2 (#caa171)
R 2 (#7807d2)
U 3 (#a77fa3)
L 2 (#015232)
U 2 (#7a21e3)"""
r=get_data(day=18, year=2023)
r=r.split('\n')
tr=[]
for rr in r:
    tr.append(rr.split())

#dir={'U':(0,-1),'R':(1,0),'D':(0,1),'L':(-1,0)}
#0 means R, 1 means D, 2 means L, and 3 means U
dir={'3':(0,-1),'0':(1,0),'1':(0,1),'2':(-1,0)}

def add(a,b,n):
    return (a[0]+b[0]*n,a[1]+b[1]*n)
    
pos=(0,0)
loop=[pos]

cnt=0
for t in tr:
    l=int(t[2][2:7],16)
    d=dir[t[2][7]]
    loop.append(add(loop[-1],d,l))
    cnt+=l

print(len(loop))
xs,ys=zip(*loop)
minx=min(xs)
miny=min(ys)
maxx=max(xs)
maxy=max(ys)

# shoelace algo
area = 0
j = len(xs)-1
for i in range(len(xs)-1):
   area += (xs[j] + xs[i]) * (ys[j] - ys[i])
   j = i
print(area)

binnen=(abs(area)//2-cnt//2)
    
minx=0
miny=0
for l in loop:
    minx=min(minx,l[0])
    miny=min(miny,l[1])    

opp=cnt+binnen
opp+1
```
